```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly
```

```python
df_rainfall = pd.read_csv('./data/過去九年降雨資料.csv')
df_rainfall
```

```python
df_rainfall[df_rainfall['Date'] == '2020-01-01']
```

```python
outerlsland = [466950,467300,467350,467620]
df_rainfall = df_rainfall[~df_rainfall['StationID'].isin(outerlsland)]
```

```python
df_rainfall['Date'] = pd.to_datetime(df_rainfall['Date'])
# df_rainfall['Year'] = df_rainfall['Date'].dt.year
# df_rainfall['dayofyear'] = df_rainfall['Date'].dt.dayofyear
df_rainfall
```

```python
df_rainfall_avg = df_rainfall.groupby('Date', as_index=False)['Precipitation'].mean()
df_rainfall_avg['year'] = df_rainfall_avg['Date'].dt.year
df_rainfall_avg['dayofyear'] = df_rainfall_avg['Date'].dt.dayofyear
df_rainfall_avg
```

```python
df_rainfall_avg[df_rainfall_avg['year'] == 2022].plot(x='dayofyear', y='Precipitation', kind='line')
```
